```python
import json

import wn
```

```python
with open("data/path_with_gaps.json", "r") as f:
    data = json.load(f)
```

```python
for elem in data:
    pwn = next(iter(elem["path"]))
    for key, value in elem["path"].items():
        if "*" in value["title"]:
            ss = wn.synset(key)
            print(
                key,
                ss.ili.id,
                ss.pos,
                ss.definition(),
                ss.lemmas(),
                [hypernym.lemmas() for hypernym in ss.hypernyms()],
                [hyponym.lemmas() for hyponym in ss.hyponyms()],
            )
```
